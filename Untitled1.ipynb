```python
xta,xte,yta,yte=TTS(x,y,test_size=0.2,random_state=2)
```

```python
def funcc(t):
    Country=['United States', 'others', 'United Kingdom', 'Canada']
    t.Country=np.where((t.Country==Country[0]) | (t.Country==Country[1]) |(t.Country==Country[2]) | (t.Country==Country[3]),t.Country,"others")
    return t
```

```python
ft=FT(funcc)
```

```python
coun=['Canada', 'United Kingdom', 'others', 'United States']
wellness_program=["Don't know", 'No', 'Yes']
care_options=['Not sure', 'No', 'Yes']
anonymity=["Don't know", 'Yes', 'No']
leave=['Very easy', 'Somewhat easy', "Don't know", 'Somewhat difficult','Very difficult']
mental_health_consequence=['No', 'Maybe', 'Yes']
phys_health_consequence=['No', 'Maybe', 'Yes']
coworkers=['Yes', 'No', 'Some of them']
supervisor=['Yes', 'No', 'Some of them']
mental_health_interview=['Maybe', 'Yes', 'No']
phys_health_interview=['No', 'Maybe', 'Yes']
mental_vs_physical=['Yes', "Don't know", 'No']
ct=CT([
    ("OE",OE(categories=[coun,care_options,wellness_program,anonymity,leave,mental_health_consequence,phys_health_consequence,coworkers,supervisor,mental_health_interview,phys_health_interview,mental_vs_physical]),["Country","care_options","wellness_program","anonymity","leave","mental_health_consequence","phys_health_consequence","coworkers","supervisor","mental_health_interview","phys_health_interview","mental_vs_physical"]),
    ("OHE",OHE(drop="first"),[2,3,4,5,6,7])
],remainder="passthrough")
```

```python
pp=PP(
[
    ("Function",ft),
    ("Column",ct),
    ("ss",SS()),
    ("rfc",RFC())
    
],verbose=True)
gscv=GSCV(pp,param_grid={},cv=5,scoring="accuracy",n_jobs=-1)
gscv.fit(xta,yta)
```

```python
gscv.best_score_,AS(yte,gscv.predict(xte))
```

```python
pp=PP(
[
    ("Function",ft),
    ("Column",ct),
    ("SS",SS()),
    ("rfc",SVC())
    
],verbose=True)
gscv=GSCV(pp,param_grid={},cv=5,scoring="accuracy",n_jobs=-1)
gscv.fit(xta,yta)
```

```python
gscv.best_score_,AS(yte,gscv.predict(xte))
```

```python
pp=PP(
[
    ("Function",ft),
    ("Column",ct),
    ("SS",SS()),
    ("rfc",SC(
    [
        ("rfc",RFC()),
        ("SVC",SVC())
    ],final_estimator=LO(),n_jobs=-1,passthrough=True))
    
],verbose=True)
gscv=GSCV(pp,param_grid={},cv=5,scoring="accuracy",n_jobs=-1)
gscv.fit(xta,yta)
```

```python
gscv.best_score_,AS(yte,gscv.predict(xte))
```

```python
pp=PP(
[
    ("Function",ft),
    ("Column",ct),
    ("SS",SS()),
])
xta=pp.fit_transform(xta)
xte=pp.transform(xte)
```

```python
model=Sequential()
model.add(Dense(15,activation="relu",input_dim=xta.shape[1],kernel_initializer="he_uniform"))
model.add(Dropout(0.4))
model.add(Dense(25,activation="relu",kernel_initializer="he_uniform"))
model.add(Dropout(0.4))
model.add(Dense(1,activation="sigmoid"))
model.summary()
```

```python
es=ES(monitor="val_loss",
    min_delta=0.000000000001,
    patience=200,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
    start_from_epoch=0,)
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(momentum=0.7,nesterov=True),loss=tf.keras.losses.binary_crossentropy,metrics=["accuracy"])
history=model.fit(xta,yta,validation_data=(xte,yte),epochs=800,callbacks=es)
```

```python
mp.plot(history.history["loss"],label="loss")
mp.plot(history.history["val_loss"],label="val_loss")
mp.legend();
```

```python
mp.plot(history.history["accuracy"],label="accuracy")
mp.plot(history.history["val_accuracy"],label="val_accuracy")
mp.legend();
```

```python
ypre=model.predict(xte)
ypre=np.where(ypre>0.5,1,0)
AS(yte,ypre)
```
